```python
import json
import jsonquery
import pathlib

data_dir = pathlib.Path.cwd()
while data_dir.name != "sandbox":
    data_dir = data_dir.parent

data_filename = data_dir / "f5" / "hub-api" / "equipments.json"
with open(data_filename) as stream:
    json_object = json.load(stream)
```

```python
with open(data_filename) as stream:
    print(stream.read(5000))
```
