# component sandbox

```python
import json

import jsonquery

with open("component.json") as stream:
    json_object = json.load(stream)

print(json.dumps(json_object, indent=4, sort_keys=True))
```

```python
# jsonquery.query(json_object, key_path, predicate=None, selector=None)

# Get all the blade numbers
result = jsonquery.query(
    json_object,
    ("component", "*", "config", "name"),
    selector=lambda name: name.replace("blade-", ""),
)

result = list(result)
print(json.dumps(result, sort_keys=True, indent=4))
```

```python
# jsonquery.query(json_object, key_path, predicate=None, selector=None)

result = jsonquery.query(
    json_object,
    ("component", "*", "properties", "property", "*"),
    predicate=lambda d: d["config"]["name"].startswith("QAT"),
    selector=lambda d: d["state"],
)

result = list(result)
print(json.dumps(result, sort_keys=True, indent=4))
```
