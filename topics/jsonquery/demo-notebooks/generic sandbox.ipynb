```python
import json
import pathlib

import jsonquery

data_dir = pathlib.Path.cwd()
while data_dir.name != "sandbox":
    data_dir = data_dir.parent

data_filename = data_dir / "f5" / "restconf" / "image.json"
with open(data_filename) as stream:
    json_object = json.load(stream)

print(json.dumps(json_object, sort_keys=True, indent=4))
```

```python
json_object.keys()
```

```python
items = jsonquery.query(
    json_object,
    (
        "partition",
        "state",
        "controllers",
        "controller",
        "*",
    ),
)

for item in items:
    print(json.dumps(item, sort_keys=True, indent=4))
    print("-" * 40)
```
